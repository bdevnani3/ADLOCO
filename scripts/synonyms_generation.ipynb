```python
synonyms = {}

for _c in CLASSES:
    synonyms[_c] = []
    for syn in wordnet.synsets(_c):
        for l in syn.lemmas():
            if l.name() != _c:
                synonyms[_c].append(l.name())
    
    
synonyms
```
